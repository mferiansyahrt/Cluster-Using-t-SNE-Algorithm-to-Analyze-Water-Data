```python
import pandas as pd

path = '/content/sample_data'
df = pd.read_csv(path + '/CleanedUp_DataSet_Orbi.csv') # Input the file here
```

```python
#%%

data = df.drop([df.columns[i] for i in range(0,7,1)] + ['Ufid_2','ISTD_matches'], axis = 1)

# Poluted Water Data

index = data['Ufid_1']
X = data.drop(['Ufid_1'],axis = 1)

X.describe()
```

```python
#%% Normalize Data

from sklearn.preprocessing import Normalizer

norm = Normalizer(norm = 'l1')
norm.fit(X)
X_transform = norm.transform(X)

sample = 0      # For seeing the data after normalized for each sample 
print(f'Unnormalized Data Sample number-{sample} \n {X.iloc[sample].values}\n')
print(f'Nnormalized Data Sample number-{sample} \n {X_transform[sample]}\n')

df_norm = pd.DataFrame(X_transform,columns = X.columns)
df_norm.describe()
```

```python
#%% t-SNE for reduce dimensionality

from sklearn.manifold import TSNE
from sklearn.mixture import GaussianMixture
import time

start = time.time()

dim = 2
tsne = TSNE(n_components = 2, 
            perplexity = 100,
            learning_rate = 300.0,
            verbose = 1,
            n_iter = 5000)
X_embedded = tsne.fit_transform(X_transform)                              # Clustering the data
print('t-SNE training done! Time elapsed : %.3f'%(time.time() - start))
```

```python
n_clusters = 26
gmm_model = GaussianMixture(n_components = n_clusters, random_state = 1)
label = gmm_model.fit_predict(X_embedded)     # Labeling the data
dftsne = pd.DataFrame(X_embedded,columns = ['c' + str(i) for i in range(dim)])
dftsne = pd.concat([index,dftsne],axis = 1)
dftsne['label'] = label
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = [14,10])

colors = ['red','green','blue','purple','cyan','yellow','navy','lime','maroon','magenta','grey','silver','lightpink','slategray',
          'steelblue','gold','mediumvioletred','lightgreen','darkslategrey','lawngreen',
          'olive','cadetblue','cadetblue','plum','yellowgreen','lightgrey']
for i in range(n_clusters):
  x = dftsne['c0'][dftsne['label'] == i]
  y = dftsne['c1'][dftsne['label'] == i]
  plt.text(gmm_model.means_[i][0], gmm_model.means_[i][1], str(i) ,fontsize = 20, color = 'black')
  plt.scatter(x,y,label = "Cluster " + str(i),alpha = 0.5,c = colors[i])

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Clustered Data')
plt.xlabel('c0')
plt.ylabel('c1')
plt.show()
```

```python
#%% t-SNE for reduce dimensionality

from sklearn.manifold import TSNE
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import time

def tsne_cluster():
  start = time.time()

  dim = 2
  tsne = TSNE(n_components = 2, 
              perplexity = 100,
              learning_rate = 300.0,
              verbose = 1,
              n_iter = 5000)
  X_embedded = tsne.fit_transform(X_transform)                              # Clustering the data
  print('t-SNE training done! Time elapsed : %.3f'%(time.time() - start))

  n_clusters = 26
  gmm_model = GaussianMixture(n_components = n_clusters, random_state = 1)
  label = gmm_model.fit_predict(X_embedded)     # Labeling the data
  dftsne = pd.DataFrame(X_embedded,columns = ['c' + str(i) for i in range(dim)])
  dftsne = pd.concat([index,dftsne],axis = 1)
  dftsne['label'] = label

  plt.figure(figsize = [14,10])

  colors = ['red','green','blue','purple','cyan','yellow','navy','lime','maroon','magenta','grey','silver','lightpink','slategray',
            'steelblue','gold','mediumvioletred','lightgreen','darkslategrey','lawngreen',
            'olive','cadetblue','cadetblue','plum','yellowgreen','lightgrey']
  for i in range(n_clusters):
    x = dftsne['c0'][dftsne['label'] == i]
    y = dftsne['c1'][dftsne['label'] == i]
    plt.text(gmm_model.means_[i][0], gmm_model.means_[i][1], str(i) ,fontsize = 20, color = 'black')
    plt.scatter(x,y,label = "Cluster " + str(i),alpha = 0.5,c = colors[i])

  plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
  plt.title('Clustered Data')
  plt.xlabel('c0')
  plt.ylabel('c1')
  plt.show()
  return dftsne

dftsne_2 = tsne_cluster()
```

**We set variable reduced_anom1 and reduced_anom2 as benchmark because we ran it first. Then, to calculate similiarity we use :**

$Similiarity = \frac{Total\ True}{Total\ Sample} \times 100\%$

**Then, we see variable reduced_anom1 and reduced_anom2_2 has the same pattern, where begins from index of 16,69,95 and so on. Therefore, we use similiarity formula based on this 2 variable**

Notes : The value might be changed if you run it another time, so it's not really reduced_anom1 == reduced_anom2_2

```python
# check index of reduction dimentionality

def idx_check(first,second):  # first == first run data, second = second run data
  if len(second) >= len(first): 
    index = []
    for idx in second.index:
      if idx not in first.index:
        index.append(idx)
    print("Index from second run data")
  elif len(first) >= len(second):
    index = []
    for idx in first.index:
      if idx not in second.index:
        index.append(idx)  
    print("Index from first run data")
  return index

# remove row by index
def remove_index(first,second,index):
  if len(second) >= len(first): 
    data = second.drop(index,axis = 0)
    print("Remove row by index on second run data")
  elif len(first) >= len(second):
    data = first.drop(index,axis = 0)
    print("Remove row by index on first run data")
  return data

def clear_idx(first,second):
  if len(first) >= len(second): 
    index = []
    for idx in first.index:
      if idx not in second.index:
        index.append(idx)
    data = first.drop(index,axis = 0)
    print("Remove row by index on first run data")
  elif len(second) >= len(first):
    index = []
    for idx in second.index:
      if idx not in first.index:
        index.append(idx)
    data = second.drop(index,axis = 0) 
    print("Remove row by index on second run data")
  return data
  
# Similiarity algorithm calculation

def similiarity(first,second,original_first,original_second):
  val = []
  for i in range(len(first['Ufid_1'])):
    val.append(first['Ufid_1'].values[i] == second['Ufid_1'].values[i])
  
  if len(first) == len(second):
    if len(original_second) >= len(original_first):
      num = len(original_second) - len(first)
      length = len(original_second)
    elif len(original_first) >= len(original_second):
      num = len(original_first) - len(second)
      length = len(original_first)
  data = np.array(val + [False]*num)
  similiarity_value = round((sum([i for i in data if i == True])/length),2)
  return similiarity_value*100
```

```python
# You don't have to understand this dirty code. It's just so I can run check_similiarity function to produce all of the similiarity value on correlated cluster in one click

def check_similiarity(reduced_anom1,reduced_anom2,reduced_anom1_2,reduced_anom2_2,threshold = 100):
  avg_anom1 = sum(reduced_anom1.index[0:30].values)/len(reduced_anom1.index[0:30].values)
  avg_anom2 = sum(reduced_anom2.index[0:30].values)/len(reduced_anom2.index[0:30].values)
  avg_anom1_2 = sum(reduced_anom1_2.index[0:30].values)/len(reduced_anom1_2.index[0:30].values)
  avg_anom2_2 = sum(reduced_anom2_2.index[0:30].values)/len(reduced_anom2_2.index[0:30].values)
  def disp(avg1,avg2,reduced1,reduced2):
    global reduced1_removedidx,reduced2_removedidx,val_similiarity 
    if abs(avg1 - avg2) < threshold:
      index = idx_check(reduced1,reduced2)
      check = []
      if len(reduced2) >= len(reduced1):
        reduced2_removedidx = remove_index(reduced1,reduced2,index)
        check.append(0)
      elif len(reduced2) <= len(reduced1):
        reduced1_removedidx = remove_index(reduced1,reduced2,index)
        check.append(1)

      if check[0] == 0: 
        reduced1_removedidx = clear_idx(reduced1,reduced2_removedidx)
      elif check[0] == 1:
        reduced2_removedidx = clear_idx(reduced1_removedidx,reduced2)

      val_similiarity = similiarity(reduced1_removedidx,reduced2_removedidx,reduced1,reduced2)
      print('-'*12)
      print(f'Similiarity for Cluster-{reduced1_removedidx["label"].value_counts().index[0]} and Cluster-{reduced2_removedidx["label"].value_counts().index[0]} : {val_similiarity} %\n')
    elif abs(avg1 - avg2) > threshold:
      index = idx_check(reduced1,reduced2)
      check = []
      if len(reduced2) >= len(reduced1):
        reduced2_removedidx = remove_index(reduced1,reduced2,index)
        check.append(0)
      elif len(reduced2) <= len(reduced1):
        reduced1_removedidx = remove_index(reduced1,reduced2,index)
        check.append(1)

      if check[0] == 0: 
        reduced1_removedidx = clear_idx(reduced1,reduced2_removedidx)
      elif check[0] == 1:
        reduced2_removedidx = clear_idx(reduced1_removedidx,reduced2)

      val_similiarity = similiarity(reduced1_removedidx,reduced2_removedidx,reduced1,reduced2)
      print('-'*12)
      print(f'Similiarity for Cluster-{reduced1["label"].value_counts().index[0]} and Cluster-{reduced2["label"].value_counts().index[0]} Has no Similiarity or less similiarity, with Similiarity of : {val_similiarity} %\n')

  disp(avg_anom1,avg_anom2_2,reduced_anom1,reduced_anom2_2)
  disp(avg_anom2,avg_anom1_2,reduced_anom2,reduced_anom1_2)
  disp(avg_anom1,avg_anom1_2,reduced_anom1,reduced_anom1_2)
  disp(avg_anom2,avg_anom2_2,reduced_anom2,reduced_anom2_2)
#  return avg_anom2,avg_anom1_2
```
